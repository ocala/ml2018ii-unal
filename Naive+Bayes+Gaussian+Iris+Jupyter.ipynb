# Naive Bayes Gaussian - Iris flower data set  - Quantopian

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

The Iris flower data set contains a set of 150 records 4 attributes Sepal Length , Sepal Width, Petal Length ,Petal Width and and fith value for  the class. 
* https://en.wikipedia.org/wiki/Iris_flower_data_set
* https://archive.ics.uci.edu/ml/datasets/iris
* http://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html
* http://scikit-learn.org/stable/modules/naive_bayes.html

```python
irisObject = datasets.load_iris()
iris_data = irisObject.data
iris_target = irisObject.target
labels = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
iris_labels = [labels[x] for x in iris_target]
#iris = np.c_[ iris_data.astype(int),  iris_labels]
#iris[:,0:3] = iris[:,0:3].astype(np.inexact)
#iris
#iris = np.array([ x[0:3] for x in iris_data])
iris = []
for index, row in enumerate(iris_data):
#    print ("%a" % row)
#    print ("%a" % iris_labels[index])
#    print ("%a" % row.tolist())
#    print (row.tolist().append(iris_labels[index]))
    row_ = row.tolist()
    iris.append([row_[0],row_[1],row_[2],row_[3],iris_labels[index]])
iris
#iris = iris.tolist()
#iris
```

```python
X = np.array([x[0:4] for x in iris])

label = [x[4] for x in iris]

y = [] 
color = [] 
for i in range(len(label)):
    if label[i] == 'Iris-setosa':
        color.append('red')
        y.append(0)
    elif label[i] == 'Iris-versicolor':
        color.append('green')
        y.append(1)
    else:
        color.append('blue')
        y.append(2)

y = np.array(y).tolist()
```

## We load the Gaussian Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

Now we are going to illustrate how the Gaussian Naive Bayes clasifier using only the first two attributes Sepal Length and Sepal Width  work

```python
x = X[:,0:2]
     
fig, ax = plt.subplots()
ax.scatter(x[:,0], x[:,1],  c=color, s=50);
ax.set_title('Naive Bayes Model', size=14)
xlim = (4, 9)
ylim = (1, 5)

xg = np.linspace(xlim[0], xlim[1], 80)
yg = np.linspace(ylim[0], ylim[1], 80)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for l, c in enumerate(['red', 'green','blue']):
    mask = (y == l)
    mu, std = x[mask].mean(0), x[mask].std(0)
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=c.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=c, alpha=0.2)
    
ax.set(xlim=xlim, ylim=ylim);
```

```python
x = X[:,0:2]
ymesh = gnb.fit(x, y)

x0 =  np.linspace(4,9,160)
x1 =  np.linspace(1,5,160)
g = np.meshgrid(x0,x1)
xmesh = np.append(g[0].reshape(-1,1),g[1].reshape(-1,1),axis=1)

ymesh = gnb.predict(xmesh)

colormesh = [] 
for i in range(len(ymesh)):
    if ymesh[i] == 0:
        colormesh.append('red')
    elif ymesh[i] == 1:
        colormesh.append('green')
    else:
        colormesh.append('blue')

   
plt.scatter(xmesh[:, 0], xmesh[:, 1], c=colormesh )
plt.title('Desicion Boundary', size=14)

y_pred = gnb.predict(x)
print "Number of mislabeled points out of a total %d points : %d" % (len(X),(y != y_pred).sum())
```
