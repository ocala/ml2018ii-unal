```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import datasets ## imports datasets from scikit-learn
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from matplotlib import rcParams

plt.style.use('seaborn')


rcParams['figure.figsize'] = [8.0, 6.0]
rcParams['figure.dpi'] = 80
rcParams['savefig.dpi'] = 100
rcParams['font.size'] = 12
rcParams['legend.fontsize'] = 'large'
rcParams['figure.titlesize'] = 'medium'


import seaborn as sns
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
data = datasets.load_boston() ## loads Boston dataset from datasets library
```

```python
print(data.DESCR)
```

```python
# define the data/predictors as the pre-set feature names  
X = pd.DataFrame(data.data, columns=data.feature_names)
X.head()
```

```python
# Put the target (housing value -- MEDV) in another DataFrame
Y = pd.DataFrame(data.target, columns=["MEDV"])
Y.head()
```

```python
plt.figure(figsize=(9, 8))
sns.distplot(Y['MEDV'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

```python
X.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); # ; avoid having the matplotlib verbose informations
```

## Only RM and LSTAT seems to have kind of the same distribution that the target. 

```python
full_df = pd.concat([X, Y], axis = 1)
full_df.head()
```

```python
df_num_corr = full_df.corr()['MEDV'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There are {} strongly correlated values with MEDV:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
rcParams['lines.markersize'] = np.sqrt(20)
for i in range(0, len(full_df.columns), 5):
    sns.pairplot(data=full_df,
                x_vars=full_df.columns[i:i+5],
                y_vars=['MEDV'])
```

```python
corr = full_df.drop('MEDV', axis=1).corr() # We already examined SalePrice correlations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
#split the data
golden_features_list
```

```python
df1 = full_df[['RM','LSTAT','PTRATIO']]
df1.head()
```

# KNN

```python
#split the data
X_train, X_test, y_train, y_test = train_test_split(df1, Y, test_size = .25, random_state=25) #25% hold out for testing
```

```python
k = 3
```

```python
# instantiate learning model (k = 3)
knn = KNeighborsRegressor(n_neighbors=k)
```

```python
knn.fit(X_train, y_train)
```

```python
# predict the response
Y_pred = knn.predict(X_test)
```

```python
mean_squared_error(y_test, Y_pred) #MSE
```

```python
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.scatter(y_test, Y_pred, color='blue', linewidth=3)
#plt.scatter(X_test, Y_predLinear, color='black')
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
```

# LinearRegression

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

Y_predLinear = lm.predict(X_test)

fig = plt.gcf()
fig.set_size_inches(15,8)
plt.scatter(y_test, Y_predLinear, color='blue', linewidth=3)
#plt.scatter(X_test, Y_predLinear, color='black')
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")

print(Y_predLinear.size)
```

```python
mean_squared_error(y_test, Y_predLinear) #MSE
```

# stat models

```python
mean_squared_error(Y, results.predict())
```
